```python
import numpy as np
import sys
import datetime
import copy
import itertools
import math

a = '9.txt'
rawmatrix= np.loadtxt (a)
matrix= np.loadtxt (a)

permutation_paths =[]
index =[]
path=[]
load_path=[]
min_array=[]
indexmatrix =[]

def count_load(index,matrix): #hitung beban tetutup (ujung ke ujung)
	sum = 0
	for i in range (0,len(index)-1):
		sum = sum+ matrix[index[i],index[i+1]]
	sum = sum + matrix[index[-1],index[0]]
	return sum

def minl(l): #nyari index dari nilai paling minimalal dalam array
  return l.index(min(l))

def bruteforce(matrix): #fungsi bruteforce
	for i in range(0,len(matrix)):
		index.append(i)
	permutation_paths = list(itertools.permutations(index, len(matrix)))

	i=0
	while i <= math.factorial(len(matrix)-1):
		path = permutation_paths[i]
		load_path.append(count_load(path,matrix))
		print("path : " + str(path) + "mempunyai count_load : " + str(count_load(path,matrix)))
		i+=1
	return permutation_paths[minl(load_path)]

############################### bruteforce ##################

loadlist=[]
pathlist=[]
start_time = datetime.datetime.now()
path = bruteforce(matrix)
end_time = datetime.datetime.now()
total_time= end_time-start_time
loadlist.append(count_load(path,matrix))
pathlist.append(path)
print(matrix)
print ("Jalur dengan path terkecil \t: " + str(pathlist[0]))
print ("Bobot Terkecil \t\t\t: "+ str(loadlist[0]))
print ("Algoritm Time \t\t\t:" + str(total_time.seconds) + " second")

#perlu dikembangkan algoritma aproximasi sendiri nanti dibandingkan dari sisi nilai dan time
```
